1. Lists  Ordered, Mutable Collections
Lists are Pythons primary sequence type for storing ordered collections of items and can contain mixed data types. Lists are mutable, meaning their content can be changed after creation.

```python
# 1. Creating a list (using square brackets [])
numbers = [1, 2, 3, 4]   # list of integers
mixed = [1, "two", 3.0, False]  # list can contain mixed data types

print("Initial List (numbers):", numbers)

# 2. Indexing (Accessing elements, index starts at 0)
print("First element (index 0):", numbers[0])        # 1
print("Last element (negative index):", numbers[-1]) # 4

# 3. Mutability (Modifying elements)
# Change the second element (index 1) from 2 to 99
numbers[1] = 99
print("List after modification (index 1 is 99):", numbers)

# 4. Adding/Removing elements
numbers.append(5)   # adds 5 to the end
print("List after append(5):", numbers)

# since 2 is no longer in the list, remove an element that exists, e.g. 99
numbers.remove(99)
print("List after remove(99):", numbers)

# 5. Common operations: Length and Slicing
print("Length of list:", len(numbers))
print("Sublist (slicing numbers[1:3]):", numbers[1:3])
```

2. Tuples  Ordered, Immutable Collections
Tuples are similar to lists as they are ordered sequences of elements. However, the key difference is that tuples are immutable; once created, their elements cannot be changed or replaced. Tuples are often used for data that shouldn't change (like coordinates) or for grouping heterogeneous data.

```python

# 1. Creating a tuple (using parentheses or just commas)
trio = ("Alice", 30, 5.5) # tuple with a string, int, and float [10]
no_paren = 1, 2, 3 # tuple packing without parentheses [10]

print("Initial Tuple (trio):", trio)

# 2. Accessing elements (similar to lists)
print("Accessing the first element:", trio)

# 3. Immutability Implication
# NOTE: The following line, if executed, would raise an error (Immutability error): [8]
# trio = "Bob"
# print(trio) # To change a tuple, you must create a new one [8].

# 4. Tuple Unpacking (A neat feature)
name, age, height = trio
print(f"Unpacked variables: Name={name}, Age={age}")
```

3. Sets  Unordered Collections of Unique Elements
Sets store unique values without any particular order, analogous to a mathematical set. They are useful for eliminating duplicates or performing fast membership checks. Sets are mutable.

```python

# 1. Creating a set (using curly braces)
# Duplicates (2) are removed during creation [11, 13]
nums = {1, 2, 2, 3}
print("Initial Set (nums):", nums) # Output will be {1, 2, 3}

# Note: The order is not preserved in a set [13]
fruits = {"apple", "banana", "cherry"}
print("Initial Set (fruits):", fruits)

# 2. Adding and Removing elements (Sets are mutable)
fruits.add("orange")
print("Set after adding 'orange':", fruits)
fruits.remove("banana")
print("Set after removing 'banana':", fruits)

# 3. Set Operations (Mathematical operations) [14]
evens = {2, 4, 6, 8}
odds = {1, 3, 5, 7, 9}
others = {4, 5, 6}

# Union (set1 | set2) - combining unique elements [14]
print("Union (evens | odds):", evens | odds)

# Intersection (&) - finding common elements [14]
print("Intersection (evens & others):", evens & others)

# Difference (-) [14]
print("Difference (odds - {3,7}):", odds - {3, 7})

# 4. Membership Test (Very fast lookup) [14]
print("Is 'apple' in fruits?", "apple" in fruits)
```

4. Dictionaries  Key-Value Mappings
Dictionaries (dict) store mappings from unique, immutable keys to values. They are incredibly important for representing structured data and performing fast lookups by key. Dictionaries are mutable.

```python

# 1. Creating a dictionary (using curly braces with key:value pairs)
person = {
    "name": "Alice",
    "age": 25,
    "city": "Dubai"
} # [15, 19]

# Another example: Country capitals [16]
capitals = {"UAE": "Abu Dhabi", "France": "Paris", "Japan": "Tokyo"}

print("Initial Dictionary (person):", person)

# 2. Accessing values (using the key)
print("Person's name:", person["name"]) # person["name"] gives "Alice" [19]
print("Capital of France:", capitals["France"]) # outputs "Paris" [16]

# 3. Adding/Modifying entries (Dictionaries are mutable)
# Updating an existing value [18]
person["age"] = 26
print("Age after modification:", person["age"])

# Adding a new entry [18]
person["email"] = "alice@example.com"
print("Dictionary after adding email:", person)

# 4. Iteration (Looping through keys and values) [18]
print("\nIterating through the person dictionary:")
for key in person:
    # Print key and its corresponding value [18]
    print(key, "->", person[key])
```

B. Manual Writing Example
This demonstrates writing data in write mode (w), which is destructive (it wipes the file clean if it already exists):

```python

# Step 1: Open the file in write mode ('w') [1, 3]
# Note: If output.txt exists, 'w' will wipe it clean [3]
file_handle = open("output.txt", "w")

# Step 2: Do the thing (e.g., use the write method) [3]
file_handle.write("Hello, World!")

# Step 3: The manual close call is mandatory [3]
# If you forget this step, your data might not even get saved properly [3]
file_handle.close()
```

2. The Modern, Pythonic Way (with Statement)
The with statement is the established, correct, professional way to work with files because it guarantees that Python automatically closes the file for you, even if your code crashes.
A. Pythonic Reading Example
This demonstrates looping through a file line by line using the with statement:


```python
# The magic of the 'with' statement [2]
# Python automatically closes the file the moment the indented block is done [2]
with open("data.txt", "r") as file: # 'file' is the temporary name/handle used [2]

    # Loop through the file line by line [2]
    for line in file:
        # Use .strip() to clean up the line, getting rid of invisible spaces or new lines [2]
        print(line.strip())

# Notice what is missing: No manual file closing [2]
```

B. Pythonic Writing Example
This demonstrates writing a list of data to a file in write mode (w), ensuring the file is safely saved and closed:

Note on Modes:
 R is for read.
 W is for write (destructive).
 A is for append (adds new stuff to the end, much safer than 'w')

```python
# 1. Start with the data (a list of numbers, as discussed in the source) [4]
numbers_list = [1,2,3,4]

# 2. Open the file in write mode ('w'), giving it the handle 'f' inside the block [4]
with open("results.txt", "w") as f:

    # Loop through our list [4]
    for num in numbers_list:
        # Convert each number to a string (f.write requires a string) [4]
        data_to_write = str(num) + "\n" # Added newline for readability

        # Write the string to the file using the write method [4]
        f.write(data_to_write)

# When the loop finishes and the indented block is left,
# the file is automatically saved and closed
```
