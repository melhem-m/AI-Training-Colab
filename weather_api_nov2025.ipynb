```python
import requests

def get_weather(city):
   url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=79007029562c2ab46617d801d8018750&units=metric"
    #Open Weather API KEY: 07b531d15baecad9e15f3cca72294746
   try:
       response= requests.get(url)
       weather_data = response.json()

       temperature = weather_data["main"]["temp"]
       realfeel = weather_data["main"]["feels_like"]
       humidity = weather_data["main"]["humidity"]
       max_temp = weather_data["main"]["temp_max"]
       min_temp = weather_data["main"]["temp_min"]
       wind_speed = weather_data["wind"]["speed"]
       country = weather_data["sys"]["country"]

       print(f"Weather in {city}:")
       print(f"Temperature {temperature}¬∞C:")
       print(f"Read Feel {realfeel}¬∞C:")
       print(f"Humidity : {humidity}%")
       print(f"Max Temperature: {max_temp}¬∞C:")
       print(f"Min Temperature: {min_temp}¬∞C:")
       print(f"Wind Speed: {wind_speed} m/s:")
       print(f"Country: {country}")

   except requests.exceptions.RequestException:
        print("Error occured")


#usage
city_name = input("Enter the name of city:")
get_weather(city_name)
```

```python
import gradio as gr
import requests

def get_weather(city):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=79007029562c2ab46617d801d8018750&units=metric"
    response = requests.get(url)
    data = response.json()
    if data["cod"] == 200:
        temp = data["main"]["temp"]
        realfeel = data["main"]["feels_like"]
        description = data["weather"][0]["description"]
        return f"""
The current temperature in {city} is {temp}¬∞C with {description}.
The Read Feel in {city} is {realfeel}¬∞C.
Stay Safe. """
    else:
        return "Sorry, I couldn't find weather information for that city."

iface = gr.Interface(fn=get_weather, inputs="text", outputs=gr.Textbox(lines=10, type="text", label="Weather Information"))
iface.launch()
```
